```python
import sys
import os
import re
import fire
from pathlib import Path


sys.path.append('/home/jeremy/Library')
from storage.cherrytree_xml import CherryTree
from document.document import Document
from utility.helpers import make_identifier, snake_case
```

```python
def story_links(doc_index, base_node):
    ct = CherryTree(doc_index)
    story_dir = Path('stories')
    
    for node in ct.nodes(base_node):
        for link in [l for l in node.links if l.type == 'file']:
            link_path = story_dir.joinpath(Path(link.href).name)
            node.remove_element(link)
            node.insert_link(href=link_path, text='Story')
    ct.save()
            
            
    
```
